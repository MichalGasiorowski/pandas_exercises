### Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv). 

### Step 3. Assign it to a variable called chipo.

```python
chipo = pd.read_csv("https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv", delimiter="\t")
```

```python
chipo["item_price"] = chipo["item_price"].str.slice(1).astype(np.float)
chipo.head(10)
```

### Step 4. How many products cost more than $10.00?

```python
chipo_filtered = chipo.drop_duplicates(['item_name', 'quantity', 'choice_description'])

#chipo_one = chipo_filtered[chipo_filtered["quantity"] == 1]
chipo_filtered["price_per_item"] = chipo_filtered["item_price"] / chipo_filtered["quantity"]
price_per_item_more_than_10 = chipo_filtered["price_per_item"] > 10
#chipo_filtered[chipo_filtered["price_per_item"] > 10]["item_name"].unique()
chipo_10 = chipo_filtered[price_per_item_more_than_10].sort_values(by="price_per_item", ascending=False)

chipo_10["item_name"].nunique()

```

### Step 5. What is the price of each item? 
###### print a data frame with only two columns item_name and item_price

```python
chipo_filtered[["item_name", "price_per_item"]].drop_duplicates("item_name").sort_values(by="price_per_item", ascending=False)
#chipo_filtered[["item_name", "item_price"]].drop_duplicates("item_name").sort_values(by="item_price", ascending=False)

```

### Step 6. Sort by the name of the item

```python
chipo_filtered.sort_values(by="item_name")
```

### Step 7. What was the quantity of the most expensive item ordered?

```python
max_price_index = chipo_filtered["item_price"].argmax()

chipo_filtered.iloc[max_price_index]
```

### Step 8. How many times was a Veggie Salad Bowl ordered?

```python
chipo_filtered[chipo_filtered["item_name"] == "Veggie Salad Bowl"].count()
```

### Step 9. How many times did someone order more than one Canned Soda?

```python
more_than_one_canned_soda = (chipo_filtered["item_name"] == "Canned Soda") & (chipo_filtered["quantity"] > 1)
len(chipo_filtered[more_than_one_canned_soda])
```
